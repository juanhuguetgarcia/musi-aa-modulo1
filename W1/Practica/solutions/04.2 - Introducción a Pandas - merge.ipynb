## Cargar datos y primera inspección

```python
import pandas as pd
```

```python
df_iris = pd.read_csv('data/iris_dataset.csv')
df_iris_recolectors = pd.read_csv('data/iris_recolectors.csv',
                                  dtype={'rec_id': str})
df_recolectors_hours= pd.read_csv('data/recolectors_hours.csv')
```

```python
df_iris
```

```python
df_iris['registry_id'].nunique()
```

```python
df_iris_recolectors
```

```python
df_iris_w_info_rec = df_iris.merge(df_iris_recolectors, on=['registry_id'])
```

```python
df_iris_w_info_rec
```

```python
df_iris_w_info_rec = df_iris.merge(df_iris_recolectors, on=['registry_id'], how='inner')
```

```python
df_iris_w_info_rec
```

```python
df_num_datapoints_per_rec = df_iris_w_info_rec.groupby('rec_id')[['iris_class']].count()
```

```python
df_num_datapoints_per_rec
```

```python
df_recolectors_hours
```

```python
df_recolectors_hours['id'] = df_recolectors_hours['id'].astype(str)
```

```python
df_recolectors_hours['id'].dtype
```

```python
df_num_datapoints_hours = df_num_datapoints_per_rec.merge(df_recolectors_hours, left_index=True, right_on='id')
```

```python
df_num_datapoints_hours['mean_recolection_time'] = df_num_datapoints_hours['iris_class'] / df_num_datapoints_hours['hours']
```

```python
df_num_datapoints_hours
```
